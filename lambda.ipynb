```python
#1. Write a list comprehension to generate squares of numbers from 1 to 10.
sq= [i*i for i in range(1,11)]
print(sq)
```

```python
#2.Create a list of even numbers between 1 and 50 using list comprehension.
even= [i for i in range(1,51) if i%2==0]
print(even)
```

```python
#3.Convert all strings in a list to uppercase using list comprehension.
lst1 = ["apple","cat","dog"]
upper_lst = [i.upper() for i in lst1]
print(upper_lst)
```

```python
#4.Given a list of integers, create a new list that contains only positive numbers.
nums = [-2,5,-1,7,3]
pos = [i for i in nums if i>0]
print(pos)
```

```python
#5. Create a list of tuples (num, num^2) for numbers 1 to 5.
tup = [(i,i*i) for i in range(1,6)]
print(tup)
```

```python
#6.Extract all vowels from a given string using list comprehension.
s = "data science"
vowels = [i for i in s if i in "aeiou"]
print(vowels)
```

```python
#7. Flatten a 2D list using list comprehension.
l = [[1,2],[3,4],[5,6]]
flat = [j for i in l for j in i]
print(flat)
```

```python
#8.Replace all negative numbers in a list with 0 using list comprehension.
num= [-3,4,-1,5]
new_num= [i if i>=0 else 0 for i in num]
print(new_num)
```

```python
#9.Given a list of words, create a list of lengths of each word.
words = ["data","science","python"]
lengths = [len(i) for i in words]
print(lengths)
```

```python
#10. Filter out words that start with the letter 'A' or 'a'.
words = ["Apple","ball","ant","cat"]
filtered = [i for i in words if not i.lower().startswith("a")]
print(filtered)
```

```python
#11..From a list of numbers, generate a list of “even” or “odd” strings.
nums = [1,2,3,4]
eo = ["even" if i%2==0 else "odd" for i in nums]
print(eo)
```

```python
#12. Create a list of numbers divisible by both 3 and 5 in range 1–100.
div = [i for i in range(1,101) if i%3==0 and i%5==0]
print(div)
```

```python
#13.Write a nested list comprehension to generate a multiplication table for 1–5.
table = [[i*j for j in range(1,6)] for i in range(1,6)]
print(table)
```

```python
#14.Convert a dictionary’s keys into a list using list comprehension.
d = {"a":1,"b":2,"c":3}
keys = [k for k in d]
print(keys)
```

```python
#15.Extract numeric digits from a string using list comprehension.
s = "a1b2c3"
digits = [i for i in s if i.isdigit()]
print(digits)
```

```python
#16.Use list comprehension to remove all spaces from a string.
s = "data science python"
no_space = [i for i in s if i!=" "]
print(no_space)
```

```python
#17.Create a list of characters that appear more than once in a string.
s = "programming"
multi = [i for i in set(s) if s.count(i)>1]
print(multi)
```

```python
#18.From a list of sentences, generate a list of all words.
sent = ["I love python","Data science is good"]
words = [j for i in sent for j in i.split()]
print(words)
```

```python
#19. Create a list of unique elements from a list using list comprehension + condition.
lst = [1,2,2,3,4,4,5]
unique = [i for i in set(lst)]
print(unique)
```

```python
#20.Generate all pairs (x, y) where x is from list A and y is from list B.
A = [1,2]
B = [3,4]
pairs = [(x,y) for x in A for y in B]
print(pairs)
```

Lambda

```python
#1.Write a lambda to add two numbers.
add = lambda a,b: a+b
print(add(2,3))
```

```python
# 2. Create a lambda to check if a number is even.
even = lambda x: x%2==0
print(even(4))
```

```python
# 3. Write a lambda to get the last character of a string.
last = lambda s: s[-1]
print(last("data"))
```

```python
#4.Use lambda with map() to square every number in a list.
nums = [1,2,3]
sq = list(map(lambda x:x*x, nums))
print(sq)
```

```python
#Q5.Use lambda with filter() to get only odd numbers from a list.
nums = [1,2,3,4,5]
odd = list(filter(lambda x:x%2!=0, nums))
print(odd)
```

```python
#6Use sorted() + lambda to sort a list of tuples by second value.
t = [(1,3),(2,1),(4,2)]
sorted_t = sorted(t, key=lambda x:x[1])
print(sorted_t)
```

```python
# 7.Create a lambda to check if a string is a palindrome.
pal = lambda s: s==s[::-1]
print(pal("madam"))
```

```python
#8.Use lambda to find maximum of three numbers.
mx = lambda a,b,c: max(a,b,c)
print(mx(2,5,3))
```

```python
#9.Write a lambda to reverse a string.
rev = lambda s: s[::-1]
print(rev("python"))
```

```python
#10.Use lambda with map() to convert a list of strings to integers.
lst = ["1","2","3"]
ints = list(map(lambda x:int(x), lst))
print(ints)
```
