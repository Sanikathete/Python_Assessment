```python
#12
salary = float(input("Enter employee salary in LPA: "))
rating = input("Enter rating (A / B / C / D): ").upper()

increment = 0

if salary <= 5:
    if rating == 'A':
        increment = 0.16
    elif rating == 'B':
        increment = 0.12
    elif rating == 'C':
        increment = 0.10
    elif rating == 'D':
        increment = 0.06

elif salary <= 10:
    if rating == 'A':
        increment = 0.14
    elif rating == 'B':
        increment = 0.10
    elif rating == 'C':
        increment = 0.08
    elif rating == 'D':
        increment = 0.06

elif salary <= 15:
    if rating == 'A':
        increment = 0.08
    elif rating == 'B':
        increment = 0.06
    elif rating == 'C':
        increment = 0.04
    elif rating == 'D':
        increment = 0

elif salary <= 23:
    if rating == 'A':
        increment = 0.07
    elif rating == 'B':
        increment = 0.05
    elif rating == 'C':
        increment = 0.04
    elif rating == 'D':
        increment = 0

else:
    print("Salary out of range")
    increment = 0

increment_amount = salary * increment
new_salary = salary + increment_amount

print("Increment Amount:", increment_amount, "LPA")
print("New Salary:", new_salary, "LPA")
```

```python
#46. Function to calculate age till now

from datetime import datetime

def calculate_age(dob):
    dob_date = datetime.strptime(dob, "%d-%m-%Y")
    today = datetime.today()
    age = today.year - dob_date.year
    print("Age:", age)

calculate_age("01-01-2000")
```

```python
#47. Check eligibility based on DOB and eligibility age

from datetime import datetime

def check_eligibility(dob, elig_age):
    dob_date = datetime.strptime(dob, "%d-%m-%Y")
    today = datetime.today()
    age = today.year - dob_date.year

    if age >= elig_age:
        print("Eligible")
    else:
        print("Not Eligible")

check_eligibility("01-01-2005", 18)
```

```python
#using oops
class Bill:
    def __init__(self):
        self.items=[]

    def getdata(self):
        try:
            n=int(input("How many items: "))
            for i in range(n):
                print("Item",i+1)
                sr=input("Enter sr no: ")
                menu=input("Enter menu: ")
                quant=input("Enter quantity: ")
                price=input("Enter price: ")
                self.items.append([sr,menu,quant,price])
        except ValueError:
            print("Invalid input")

    def showbill(self):
        print("="*80)
        hotel_name="Welcome to Sai Anand"
        print("{:^80}".format(hotel_name))
        print("="*80)
        print("{0:5}{1:^25}{2:>10}{3:>20}".format("Sr","Menu","Quant","Price"))

        total=0
        for i in self.items:
            print("{0:5}{1:^25}{2:>10}{3:>20}".format(i[0],i[1],i[2],i[3]))
            try:
                total=total+(int(i[2])*int(i[3]))
            except ValueError:
                print("Invalid quantity or price")

        print("-"*80)
        print("Total :",total)

b=Bill()
b.getdata()
b.showbill()
```

```python
class CollegeBill:
    def __init__(self):
        self.subject=""
        self.analytics="N"
        self.hostel="N"
        self.food_months=0
        self.transport=""
        self.total=0

    def getdata(self):
        try:
            self.subject=input("Enter subject (HR/Finance/Marketing/DS): ")
            if self.subject!="DS":
                self.analytics=input("Analytics Y/N: ").upper()
            else:
                self.analytics="N"

            self.hostel=input("Hostel Y/N: ").upper()
            self.food_months=int(input("Food months: "))
            self.transport=input("Transport (semester/annual): ").lower()
        except:
            print("Invalid input")

    def calculate(self):
        course_fee=200000

        if self.analytics=="Y":
            course_fee=course_fee+(200000*0.10)

        hostel_fee=0
        if self.hostel=="Y":
            hostel_fee=200000

        food_cost=self.food_months*2000

        if self.transport=="semester":
            transport_cost=13000
        elif self.transport=="annual":
            transport_cost=26000
        else:
            transport_cost=0

        self.total=course_fee+hostel_fee+food_cost+transport_cost

    def showbill(self):
        print("="*60)
        print("COLLEGE BILL")
        print("="*60)
        print("Subject :",self.subject)
        print("Analytics :",self.analytics)
        print("Hostel :",self.hostel)
        print("Food Cost :",self.food_months*2000)
        print("Transport :",self.transport)
        print("-"*60)
        print("Total :",self.total)

c=CollegeBill()
c.getdata()
c.calculate()
c.showbill()
```
